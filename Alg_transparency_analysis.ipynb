## Analysis based on types of surveillance

```python
alg_df
```

```python
ind_surv_df = alg_df[["agency", "year", "pop_individuals"]]
```

```python
ind_surv_df
```

```python
yes_df = ind_surv_df[ind_surv_df['pop_individuals'] == 'yes']

# Count 'yes' responses for each agency
yes_counts = yes_df.groupby('agency').size()

# Sort counts in descending order and remove agencies with zero counts
yes_counts = yes_counts.sort_values(ascending=False)
yes_counts = yes_counts[yes_counts > 0]  # Remove agencies with zero counts

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Use the viridis color map scaled to the number of agencies with non-zero counts
colors = plt.cm.viridis(np.linspace(0, 1, len(yes_counts)))

# Bar plot
yes_counts.plot(kind='bar', color=colors, ax=ax)

ax.set_xlabel('Agency')
ax.set_ylabel('Count of algorithms with personal data')
ax.set_title('Which Agencies use the most personal data for algorithms?')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
plt.savefig("personal_data_use_alg.png")
```

```python
surv_df = pd.DataFrame(alg_df[['agency','pop_geographic_space',
       'pop_FDNY_employees', 'pop_property', 'pop_individuals',
       'pop_civilian_fatalities', 'pop_group_org_business',
       'pop_other_species']])
```

```python
surv_df
```

```python
surv_df["pop_property"].value_counts()
```

```python
# Filter columns that start with 'pop_'
pop_columns = [col for col in surv_df.columns if col.startswith('pop_')]

# Reshape the DataFrame
melted_df = pd.melt(surv_df, id_vars=['agency'], value_vars=pop_columns, var_name='category', value_name='response')

# Filter for 'yes' responses only
yes_df = melted_df[melted_df['response'] == 'yes']

# Count 'yes' responses for each category by agency
yes_counts = yes_df.groupby(['agency', 'category']).size().unstack(fill_value=0)

# Sort the DataFrame by total 'yes' responses in descending order
sorted_yes_counts = yes_counts.sum(axis=1).sort_values(ascending=False)
yes_counts = yes_counts.loc[sorted_yes_counts.index]

# Plotting
fig, ax = plt.subplots(figsize=(10, 8))
bottom = np.zeros(len(yes_counts))
ax.grid(color="grey", linewidth=.2)

#colors = plt.cm.viridis(np.linspace(0, 1, len(pop_columns)))  # Generating a color map
colors=['#377eb8', '#ff7f00', '#4daf4a', '#008080', '#984ea3', '#ffdb58', '#a6cee3']

for i, column in enumerate(yes_counts.columns):
    ax.bar(yes_counts.index, yes_counts[column], bottom=bottom, label=column, color=colors[i])
    bottom += yes_counts[column].values

custom_labels = ['FDNY Employees', 'Civilian Deaths', 'Geographic Space', 'Organization/Business', 'Individuals', 'Animals', 'Property']

ax.set_xlabel('Agency')
ax.set_ylabel('Count of algorithms using each data type')
ax.set_title('What data do public sector algorithms use?')
ax.set_ylim(top=33)
ax.legend(custom_labels, title='Type of Data')

plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("dtypes_public_algs.png", dpi=500)
```

```python
# Filter columns that start with 'use_'
use_columns = [col for col in alg_df.columns if col.startswith('use_')]

# Reshape the DataFrame
melted_df = pd.melt(alg_df, id_vars=['agency'], value_vars=use_columns, var_name='category', value_name='response')

# Filter for 'yes' responses only
yes_df = melted_df[melted_df['response'] == 'yes']

# Count 'yes' responses for each category by agency
yes_counts = yes_df.groupby(['agency', 'category']).size().unstack(fill_value=0)

# Sort the DataFrame by total 'yes' responses in descending order
sorted_yes_counts = yes_counts.sum(axis=1).sort_values(ascending=False)
yes_counts = yes_counts.loc[sorted_yes_counts.index]

# Plotting
fig, ax = plt.subplots(figsize=(10, 8))
bottom = np.zeros(len(yes_counts))
ax.grid(color="grey", linewidth=.2)

#colors = plt.cm.viridis(np.linspace(0, 1, len(use_columns)))  # Generating a color map
colors = ['#377eb8', '#ff7f00', '#4daf4a', '#008080', '#984ea3', '#ffdb58', '#a6cee3']

for i, column in enumerate(yes_counts.columns):
    ax.bar(yes_counts.index, yes_counts[column], bottom=bottom, label=column, color=colors[i])
    bottom += yes_counts[column].values

custom_labels = ["Computer Vision", "Crime Classification", "Matching", "Optimization", "Other", "Predictive Modeling", "Speech Language Processing"]

ax.set_xlabel('Agency')
ax.set_ylabel('Number of different algorithm types')
ax.set_title('How public agencies use data: type of algorithms by agency')
ax.legend(custom_labels, title='Type of algorithm')

plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("alg_type_by_agency.png", dpi=500)

```

```python
year_df = pd.DataFrame(alg_df[["agency", "date_use"]])
```

```python
year_df.dtypes
```

```python
# Convert 'date_use' to datetime
year_df['date_use'] = pd.to_datetime(year_df['date_use'])

# Count the occurrences for each agency per year
count_df = year_df.groupby(['agency', year_df['date_use'].dt.to_period("M")]).size().reset_index(name='count')

# Sorting the DataFrame by year to ensure the plot orders the years correctly
count_df.sort_values('date_use', inplace=True)

# Convert date_use to string for labeling purposes
count_df['date_use'] = count_df['date_use'].dt.strftime('%Y-%m')

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.grid(color="grey", linewidth=.2)

# Apply a modified viridis color map, starting from a darker point in the palette
colors = plt.cm.viridis(np.linspace(0.2, 1, len(count_df['date_use'].unique())))  # Start from 20% into the colormap

# Mapping unique dates to colors
date_colors = {date: colors[i] for i, date in enumerate(count_df['date_use'].unique())}

# Scatter plot with adjusted viridis colors
for i, row in count_df.iterrows():
    ax.scatter(row['agency'], row['date_use'], s=row['count']*10, color=date_colors[row['date_use']], alpha=1, edgecolors='black', linewidth=.3)

# Customizing ticks for clarity
ax.set_xticks(count_df['agency'].unique())
unique_dates = count_df['date_use'].unique()
ax.set_yticks(unique_dates[::6])  # Adjust the step to show fewer ticks
ax.set_yticklabels(unique_dates[::6], rotation=45)  # Rotate for better readability

ax.set_xlabel('Agency')
ax.set_ylabel('Date')
ax.set_title('Date of algorithm adoption')

plt.show()

plt.savefig("y_m_date_alg_adopted.png")
```
